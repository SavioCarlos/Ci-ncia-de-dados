```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

df = pd.read_csv('dados_sinteticos_com_erros_atualizado.csv')
df.head()
```

```python
# Arredondando as colunas 'Valor' e 'Total' para duas casas decimais
df['valor'] = df['valor'].round(2)
df['total'] = df['total'].round(2)

# Preenchendo valores ausentes na coluna 'Cores' com 'Indefinido'
df['cor'].fillna('Indefinido', inplace=True)

# Substituindo os valores da coluna 'Pais' que não sejam 'Brazil' por 'Indefinido'
df.loc[df['pais'] != 'Brasil', 'pais'] = 'Brasil'

df = df.drop(columns=['cidade'], axis= 1)
df = df.drop(columns=['valor'], axis= 1)
df = df.drop(columns=['valor_y'], axis= 1)
df = df.drop(columns=['total'], axis= 1)
df = df.drop(columns=['categoria_y'], axis= 1)
df = df.drop(columns=['categoria'], axis= 1)

# Visualizando o DataFrame resultante
print(df)
df
```

```python

# Removendo duplicatas
print(df.duplicated().sum())
df = df.drop_duplicates()

```
